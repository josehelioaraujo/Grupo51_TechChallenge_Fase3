## Uso das Libs Unsloth e Transformers para Fine-Tuning de LLMs

### Unsloth

- Unsloth é uma biblioteca otimizada para fine-tuning de modelos de linguagem, especialmente focada em melhorar a eficiência e reduzir o consumo de recursos.

#### Características principais:
1. **Otimização de Memória**: Utiliza técnicas avançadas para reduzir o uso de memória GPU.
2. **Quantização**: Suporta quantização de 4-bit, permitindo o treinamento de modelos maiores em hardware mais limitado.
3. **Velocidade**: Oferece aceleração significativa no processo de fine-tuning comparado a métodos tradicionais.
4. **Compatibilidade**: Projetado para trabalhar bem com modelos populares como LLaMA e Mistral.

#### Vantagens:
- Permite fine-tuning de modelos grandes em hardware consumer-grade.
- Reduz significativamente o tempo e custo de treinamento.
- Mantém a qualidade do modelo enquanto otimiza recursos.

### Transformers

- Transformers é uma biblioteca popular da Hugging Face que fornece arquiteturas de estado da arte para NLP.

#### Características principais:
1. **Versatilidade**: Suporta uma ampla gama de modelos e tarefas de NLP.
2. **Pré-processamento**: Oferece ferramentas robustas para tokenização e preparação de dados.
3. **Integração**: Bem integrado com o ecossistema PyTorch e TensorFlow.
4. **Comunidade**: Grande comunidade e muitos modelos pré-treinados disponíveis.

#### Vantagens:
- API unificada para diferentes arquiteturas de modelos.
- Facilita o carregamento e uso de modelos pré-treinados.
- Oferece ferramentas para fine-tuning, embora não tão otimizadas quanto Unsloth.
  

-  Unsloth + Transformers
   - Prós:
     - Combina a eficiência do Unsloth com a versatilidade do Transformers.
     - Permite fine-tuning eficiente em hardware limitado.
     - Mantém a familiaridade da API do Transformers.
   - Contras:
     - Pode ser mais complexo de configurar inicialmente.
     - Limitado aos modelos suportados pelo Unsloth.


### Considerações para Fine-Tuning de LLMs

1. **Tamanho do Dataset**: Unsloth é eficiente para datasets menores a médios. Para datasets muito grandes, soluções distribuídas podem ser necessárias.

2. **Customização**: Se você precisa de alto grau de customização no processo de treinamento, uma abordagem mais manual com PyTorch pode ser preferível.

3. **Recursos Computacionais**: Unsloth brilha em cenários com GPU única ou recursos limitados. Para clusters de múltiplas GPUs, considere DeepSpeed ou outras soluções distribuídas.

4. **Tempo de Desenvolvimento**: Unsloth + Transformers pode acelerar significativamente o tempo de desenvolvimento e experimentação, especialmente para prototipagem rápida.

# Parâmetros de Configuração do Modelo

## `max_seq_length = 2048`

- Define o comprimento máximo da sequência que o modelo pode processar de uma vez.
- Valor 2048 significa que o modelo pode lidar com sequências de até 2048 tokens em uma única passagem.
- Determina o tamanho máximo do contexto que o modelo pode considerar ao gerar ou processar texto.
- Sequências mais longas permitem mais contexto, mas aumentam o uso de memória e o tempo de processamento.

## `dtype = None`

- Refere-se ao tipo de dados (data type) usado para representar os pesos do modelo.
- `None` permite que a biblioteca ou o modelo escolha o tipo de dados padrão ou mais apropriado.
- Geralmente resulta no uso de float32 (precisão simples) ou float16 (meia precisão).
- Afeta a precisão dos cálculos, o uso de memória e a velocidade de execução do modelo.

## `load_in_4bit = True`

- Indica que o modelo deve ser carregado usando quantização de 4 bits.
- Técnica de compressão que reduz significativamente o tamanho do modelo em memória.
- Usa apenas 4 bits para representar cada peso do modelo.
- Reduz drasticamente o uso de memória, permitindo que modelos grandes sejam executados em hardware com menos RAM.
- Oferece um bom equilíbrio entre eficiência de memória e qualidade de saída do modelo.

### Implicações das Configurações

Estas configurações, quando usadas em conjunto, indicam que o modelo está configurado para:

1. Processar sequências de até 2048 tokens.
2. Usar o tipo de dados padrão ou recomendado para os pesos do modelo.
3. Carregar o modelo em um formato altamente comprimido (4 bits por peso) para maximizar a eficiência de memória.

Esta combinação é particularmente útil para trabalhar com recursos computacionais limitados, mantendo a capacidade de utilizar um modelo de linguagem grande e capaz.

```python
max_seq_length = 2048
dtype = None
load_in_4bit = True
```

## Carregando modelo sem o processo de Fine-Tuning

<br>
Carregando um modelo com os mesmo parâmetros, que não será feito o processo de Fine-Tuning. <br>
Ele será usado a nível de comparação entre um modelo com e sem treinamento.

```python
modelClean, tokenizerClean = FastLanguageModel.from_pretrained(
    model_name = "unsloth/llama-3-8b-bnb-4bit",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)

FastLanguageModel.for_inference(modelClean)
```

## Geração de Respostas:

```python
command = "Using your knowledge base talk about this prodcut. If you can’t find relevant information to answer the question, respond with, “Sorry, I don’t have any information on this question.”"

alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""
```

```python
contentInput = input("Enter your question ...")
# contentInput = "the prophet"

inputs = tokenizerClean(
[
    alpaca_prompt.format(
        command,
        contentInput,
        ""
    )
], return_tensors = "pt").to("cuda")

text_streamer = TextStreamer(tokenizerClean)
_ = modelClean.generate(**inputs, streamer = text_streamer, max_new_tokens = 128)
```
