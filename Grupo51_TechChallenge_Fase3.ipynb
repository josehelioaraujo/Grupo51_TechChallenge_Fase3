# **Introdução Teórica do Fine-Tuning**

### **Funcionamento Básico do Processo de Fine-Tuning**
image.png


 O processo de fine-tuning de um LLM consiste em duas etapas principais:

- **Pré-treinamento**: Dados de texto bruto são usados para treinar um modelo inicial, criando um LLM pré-treinado com conhecimento geral.

- **Fine-Tuning** (Ajuste fino): O LLM pré-treinado é refinado usando uma base de conhecimento personalizada, resultando em um LLM ajustado para tarefas ou domínios específicos.

- Este processo permite que um modelo de linguagem genérico seja adaptado para aplicações particulares, melhorando seu desempenho em áreas específicas.
  

## **Etapas do processo**:
1. **Pré-treinamento (Pre-training)**
2. **Fine-tuning (Ajuste fino)**

## **Componentes**:
- **Raw Text Data** (Dados de Texto Bruto): Representado por ícones de cubos conectados
- **Pre-trained LLM** (LLM Pré-treinado): Representado por um ícone de documento
- **Custom Knowledge Base** (Base de Conhecimento Personalizada): Representado por um ícone de engrenagem
- **Fine-tuned LLM** (LLM Ajustado): Representado por um ícone de documento com uma pequena engrenagem

## **Fluxo do processo**:
1. Os dados de texto bruto passam pelo processo de pré-treinamento para criar o LLM pré-treinado(no nosso contexto, usamos o modelo **LLaMA 3 8B**)

2. O LLM pré-treinado, juntamente com a base de conhecimento personalizada, passa pelo processo de fine-tuning

3. O resultado final é o LLM ajustado (fine-tuned)

   
## **Aplicações em Cenários Reais**

O fine-tuning é amplamente utilizado em diversas aplicações, como:

- **Assistentes Virtuais**: Adaptar modelos para entender e responder a perguntas específicas de um domínio.
- **Análise de Sentimentos**: Especializar modelos para identificar emoções em textos.
- **Tradução Automática**: Melhorar a precisão de traduções em contextos específicos.
- **Detecção de Fraudes**: Identificar padrões suspeitos em transações financeiras.

## **Comparação com Outras Abordagens**

Comparado a outras abordagens, como o treinamento do zero , o uso de modelos pré-treinados sem ajustes, o fine-tuning oferece um equilíbrio entre eficiência e especialização:

- **Treinamento do Zero**: Requer grandes quantidades de dados e recursos computacionais, mas pode ser altamente especializado.
- **Uso de Modelos Pré-Treinados**: Rápido e eficiente, mas pode não ser suficientemente especializado para tarefas específicas.
- **Fine-Tuning**: Aproveita o conhecimento prévio do modelo e o adapta para tarefas específicas, oferecendo uma boa relação custo-benefício.

- Também é possível, comparar o Fine-Tuning com a abordadem utilizada no RAG(Retrieval-Augmented Generation (RAG), ou até mesmo, ambos podem ser usados na solução deste problema( o que seria uma melhoria no futuro deste projeto).

## **O que é RAG?**

- **Retrieval-Augmented Generation (RAG)** é uma técnica que combina modelos de linguagem grandes (LLMs) pré-treinados com fontes de dados externas para gerar respostas mais precisas e informadas. Em vez de depender apenas do conhecimento armazenado no modelo, o RAG busca informações relevantes em bases de dados externas durante o processo de geração de texto.
- Isso permite que o modelo forneça respostas atualizadas e específicas, mesmo em tópicos que não foram amplamente cobertos durante o treinamento inicial.
-  Combina a geração de texto com a recuperação de informações de fontes externas.

## **Comparação Fine-Tuning X RAG**

1. **Fine-Tuning**
   - **Vantagens**: Pode melhorar significativamente a precisão do modelo em tarefas específicas.
   - **Desvantagens**: Requer um conjunto de dados grande e relevante para a tarefa, além de ser computacionalmente intensivo;

2. **RAG**

  - **Vantagens**: Fornece respostas mais atualizadas e precisas, especialmente em tópicos dinâmicos ou menos cobertos durante o treinamento.

 - **Desvantagens**: Pode ser mais complexo de implementar, pois requer a integração de sistemas de recuperação de informações com modelos de geração de texto


- Em resmo, enquanto o ajuste fino é excelente para especializar um modelo em uma tarefa específica, o RAG é ideal para situações onde é necessário acessar informações atualizadas e variadas.
- A escolha entre essas técnicas depende do contexto e das necessidades específicas da aplicação.
   
## **Métricas de Avaliação**

Para medir a eficiência de um modelo treinado, são utilizadas várias métricas e métodos de avaliação:

 - **BLEU Score**: Mede a similaridade entre o texto gerado e a referência. Um valor mais alto indica maior similaridade.
- **ROUGE-L F1**: Avalia a qualidade da sumarização, considerando a sobreposição de n-gramas entre o texto gerado e a referência.
- **BERTScore F1**: Mede a similaridade semântica entre o texto gerado e a referência.
- **Diversidade Léxica**: Mede a variedade de palavras usadas no texto gerado.
- **Comprimento Médio**: Calcula o comprimento médio das respostas geradas.


   

## **Porque selecionamos o modelo unsloth/llama-3-8b-bnb-4bit ?**

 O modelo **unsloth/llama-3-8b-bnb-4bit** é uma versão otimizada e quantizada do LLaMA (Large Language Model Meta AI), especificamente projetada para uso eficiente em recursos computacionais limitados, o que se encaixa perfeitamente no nosso contexto. Vamos descrever melhor a nomeclatura deste modelo.

  - **unsloth**: Refere-se ao projeto Unsloth, que se concentra em otimizar modelos de linguagem para execução mais rápida e eficiente.

  - **llama-3**: Indica que este é um modelo baseado na terceira iteração da arquitetura LLaMA desenvolvida pela Meta AI (anteriormente Facebook AI Research).

  - **8b**: Significa que o modelo tem aproximadamente 8 bilhões de parâmetros. Isto o coloca na categoria de modelos de linguagem de grande escala, embora seja menor que alguns dos maiores modelos disponíveis (que podem ter centenas de bilhões de parâmetros).

  - **bnb**: Significa "bitsandbytes", uma biblioteca de quantização que permite a compressão eficiente de modelos de aprendizado profundo.

  - **4bit**: Indica que o modelo foi quantizado para usar apenas 4 bits por peso, em vez dos tradicionais 32 ou 16 bits usados em modelos de precisão total ou meia precisão.

### **Principais características e vantagens**:

 - **Eficiência de memória**: A quantização de 4 bits significa que o modelo ocupa significativamente menos espaço em memória comparado à sua versão original, permitindo seu uso em hardware com recursos mais limitados.

 - **Velocidade**: Modelos quantizados geralmente são mais rápidos para inferência, pois requerem menos operações de ponto flutuante.

 - **Capacidade preservada**: Apesar da compressão, esses modelos geralmente mantêm uma grande parte de sua capacidade de geração de texto e compreensão de linguagem.

 - **Adequado para fine-tuning**: Este modelo é particularmente adequado para técnicas de fine-tuning eficientes, como LoRA (Low-Rank Adaptation), que você está usando em seu código.

 - **Balanceamento de desempenho e eficiência**: Com 8 bilhões de parâmetros, este modelo oferece um bom equilíbrio entre capacidade de linguagem e requisitos computacionais.

### **Limitações potenciais**:

 -  **Precisão reduzida**: A quantização pode levar a uma pequena degradação na qualidade das saídas em comparação com o modelo de precisão total.

 - **Tarefas específicas**: Pode não ser ideal para tarefas que requerem precisão numérica extrema.

 - **Compatibilidade**: Nem todas as operações ou bibliotecas de aprendizado profundo são otimizadas para trabalhar com modelos quantizados de 4 bits.

### **Em Resumo**:

A razão da escolha deste modelo, pois se encaixa perfeitamente, no nosso contexto, e é  particularmente adequado para fine-tuning em tarefas específicas, possuindo um desempenho eficiente, mesmo com restrições de recursos computacionais, pois precisamos fazer inferências rápidas  e testes interativos, no ambiente do Google Colab, muito embora, tivemos que adquirir unidades computacionais, a fim de rodar todo o processo usando GPU ou invés de CPU, de modo a reduzir o tempo de execução.
